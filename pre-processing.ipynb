```python
from glob import glob
import pandas as pd
import numpy as np
```

```python
def handle_missing_values(df):
    df['x']=df['x'].replace(0, np.nan)
    df['y']=df['y'].replace(0, np.nan)
    df['z']=df['z'].replace(0, np.nan)
    return df
```

```python
csv_file_list = glob("train/*")
```

```python
df = pd.read_csv(csv_file_list[0])
df.plot()
```

# insert nan to 0.0( need to handle missing values)

```python
df_handled_missing = handle_missing_values(pd.read_csv(csv_file_list[0]))
df_handled_missing.isna().head()
```

```python
df_handled_missing.plot()
```

# remove missing values

```python
df_removed_missing = df_handled_missing.dropna()
```

```python
df_removed_missing.plot()
```

# interpolation

```python
df_interpolated = df_handled_missing.interpolate()
```

```python
df_interpolated.plot()
```
